## cv.line(	img, pt1, pt2, color[, thickness[, lineType[, shift]]]	)

画线，画线的类型可参考[LineTypes](https://docs.opencv.org/3.4.2/d0/de1/group__core.html#gaf076ef45de481ac96e0ab3dc2c29a777)

**Parameters**：
- img	Image.
- pt1	First point of the line segment.
- pt2	Second point of the line segment.
- color	Line color.
- thickness	Line thickness. 厚度
- lineType	Type of the line. See [LineTypes](https://docs.opencv.org/3.4.2/d0/de1/group__core.html#gaf076ef45de481ac96e0ab3dc2c29a777).
- shift	Number of fractional bits in the point coordinates.


```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('jpg.jpg')

img_line = cv.line(img,(0,0),(100,100),(255,0,0),5)
img_line_rgb = cv.cvtColor(img_line, cv.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.subplot(131)
plt.imshow(img_line_rgb)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis

img_line = cv.line(img,(0,0),(100,100),(255,0,0),50)
img_line_rgb = cv.cvtColor(img_line, cv.COLOR_BGR2RGB)
plt.subplot(132)
plt.imshow(img_line_rgb)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis

# 这里可以发现虽然返回用另一个变量保存新图像，但是依然改变了原图img，这是因为cv.line()函数只是返回
# 的引用类型，而没有copy data
img_line = cv.line(img,(0,0),(100,100),(255,0,255),1, cv.LINE_AA)
img_line_rgb = cv.cvtColor(img_line, cv.COLOR_BGR2RGB)
plt.subplot(133)
plt.imshow(img_line_rgb)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

## cv.polylines(	img, pts, isClosed, color[, thickness[, lineType[, shift]]]	)

绘制多边形

```python
pts = np.array([[100,5],[200,30],[100,20],[200,10]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(img_line_rgb,[pts],False,(255,0,0))

plt.imshow(img_line_rgb)
plt.show()
```

## cv.putText(	img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]	)

绘制文字，字体见[FontType](https://docs.opencv.org/3.4.2/d0/de1/group__core.html#ga0f9314ea6e35f99bb23f29567fc16e11)

**Parameters**:
- img	Image.
- text	Text string to be drawn.
- org	Bottom-left corner of the text string in the image.
- fontFace	Font type, see HersheyFonts.
- fontScale	Font scale factor that is multiplied by the font-specific base size.
- color	Text color.
- thickness	Thickness of the lines used to draw a text.
- lineType	Line type. See LineTypes
- bottomLeftOrigin	When true, the image data origin is at the bottom-left corner. Otherwise, it is at the top-left corner.


```python
cv.putText(img_line_rgb,'Hi~',(20,100), cv.FONT_HERSHEY_SIMPLEX, 4,(0,0,0),2,cv.LINE_AA)

plt.imshow(img_line_rgb)
plt.show()
```
