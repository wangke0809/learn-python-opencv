## cv.getTrackbarPos(trackbarname, winname)

Returns the trackbar position.


````python
import numpy as np
import cv2 as cv

def call_back(data):
    # 当移动滑动条时触发回调函数，入参为滑动条当前pos
    print("call back func", data)
    

cv.namedWindow('image')
cv.createTrackbar('B','image',0,255,call_back)
cv.createTrackbar('G','image',0,255,call_back)
cv.createTrackbar('R','image',0,255,call_back)

img = np.zeros((300,512,3), np.uint8)

while(1):
    cv.imshow('image',img)
    k = cv.waitKey(1) & 0xFF
    if k == 27: # ESC
        break
    r = cv.getTrackbarPos('R','image')
    g = cv.getTrackbarPos('G','image')
    b = cv.getTrackbarPos('B','image')
    
    img[:] = [b,g,r]
    
cv.destroyAllWindows()
````
